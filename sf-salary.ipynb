```python
# Find out what is the avg amount of BasePay in the dataset
salary['BasePay']=pd.to_numeric(salary['BasePay'],errors='coerce')
avg_BasePay=salary['BasePay'].mean()
avg_BasePay
```

```python
# Find out what is the highest amount of OvertimePay in the dataset
salary['OvertimePay']=pd.to_numeric(salary['OvertimePay'],errors='coerce')
max_OvertimePay=salary['OvertimePay'].max()
max_OvertimePay
```

```python
# Find out the job title of Joseph Driscoll
salary[salary['EmployeeName']=='JOSEPH DRISCOLL']['JobTitle']
```

```python
# Find out how much Joseph Driscoll makes (including benefits)
salary[salary['EmployeeName']=='JOSEPH DRISCOLL']['TotalPayBenefits']
```

```python
# Find out the name of the highest paid person (including benefits)
salary[salary['TotalPayBenefits']==salary['TotalPayBenefits'].max()]['EmployeeName']
```

```python
# Find out the name of the lowest paid person (including benefits)
salary[salary['TotalPayBenefits']==salary['TotalPayBenefits'].min()]['EmployeeName']
```

```python
# Find out the average BasePay of all employees per year between 2011 and 2014
salary[(salary['Year']>=2011) & (salary['Year']<=2014)]==salary['BasePay'].mean()
```
