```python
# display the correlation heatmap of the dataframe
sns.heatmap(auto.corr(),cmap='coolwarm',annot=True)
```

```python
# plot a line chart of the mean by Cylinders
auto.groupby('cylinders').mean().plot(kind='line')
plt.title('mean by cylinders')
plt.show()
```

```python
# plot a bar chart of the mean by mpg
auto.groupby('origin').mean().plot(kind='bar')
plt.title('mean by origin')
plt.show()
```

```python
# plot a pie chart of the mean by mpg
auto.groupby('origin').mean().plot(kind='box')
plt.title('mean by origin')
plt.show()
```

```python
# create a pair plot of mpg, weight, and origin with hue set to origin
sns.pairplot(auto[['mpg','weight','origin']],hue='origin',height=4)
```

```python
# replace the question mark with a numeric value (e.g., 0)
auto['horsepower'] = auto['horsepower'].replace('?', 0)

# convert the horsepower column to integer values
auto['horsepower'] = auto['horsepower'].astype(int)
```

```python
# filter the dataframe to include only rows where acceleration > 15 and horsepower < 170
filtered_auto =auto[(auto['acceleration']>15) & (auto['horsepower']<= 170)]
print(filtered_auto)
```

```python
# group the dataframe by cylinders and count the number of unique values in each group
auto = auto.groupby('cylinders')
counts = auto.size()
print(counts)
```

```python
# group the dataframe by model_year and count the number of rows in each group
model_year_count= auto.groupby('model_year').size()
print(model_year_count)
```
